1. 키보드로 세 수를 입력받고,입력받은 세 수를 리스트에 저장하시오. 그 후 리스트에 저장된 수 중 가장 큰 수와 가장 작은 수의 차를 출력하시오.

```python
#1번 문제 풀이 작성
numbers = []

for i in range(1, 4):
  num = int(input(f'{i}번째 수 입력: '))
  numbers.append(num)

print(f'\n입력한 수 : {numbers}')
print(f'최댓값: {max(numbers)}, 최솟값: {min(numbers)}')
print(f'차이: {max(numbers) - min(numbers)}')
```

6. 아래 딕셔너리 데이터에 stu_grade라는 키를 추가하고, 해당 키의  데이터로  점수(stu_score)에 따른 등급 데이터를 저장하시오.  
90 <= 점수      : A등급  
80 <= 점수 < 90 : B등급  
70 <= 점수 < 80 : C등급  
      점수 < 70 : D등급 

```python
class_room_data = {
    'stu_no' : [1, 2, 3, 4, 5],
    'stu_name' : ['김자바', '이자바', '박자바', '최자바', '정자바'],
    'stu_score' : [85, 65, 70, 80, 95]
}

#6번 문제 풀이 작성

grades = []

for i in range(len(class_room_data['stu_score'])):
  score = class_room_data['stu_score'][i]

  if score >= 90:
    grades.append('A등급')
  elif score >= 80:
    grades.append('B등급')
  elif score >= 70:
    grades.append('C등급')
  else:
    grades.append('D등급')

class_room_data['stu_grade'] = grades
class_room_data
```

7. 아래 데이터는 자바반과 자반에 소속된 학생들의 정보이다.  
나이가 30살 이상인 학생들의 점수의 합을 출력하세요.

```python
javaClass= {
  'calssName' : '자바반',
  'teacher' : '김자바',
  'studentList' : [
    {'name' : '홍길동', 'age': 31, 'score' : 85},
    {'name' : '이순신', 'age': 35, 'score' : 95},
    {'name' : '유관순', 'age': 29, 'score' : 80},
    {'name' : '장영실', 'age': 37, 'score' : 70},
    {'name' : '신사임당', 'age': 25, 'score' : 65},
  ]
}

# 7번 문제 답안 작성

sum1 = 0

for e in range(len(javaClass['studentList'])):
  if javaClass['studentList'][e]['age'] >= 30:
    sum1 = sum1 + javaClass['studentList'][e]['score']

sum1
```

8. 2012년 기준 대한민국에서 25세 미만의 인구수 중 인구가 가장 많은 연령대를 출력하시오.

```python
# 8번 문제 답안 작성

sum_dict = {}

for key, value in population.items():
  if isinstance(value, list):
    sum_dict[key] = sum(v for v in value if isinstance(v, (int, float)))
  elif isinstance(value, (int, float)):
    sum_dict[key] = value
sum_dict

if sum_dict:
  max_key = max(sum_dict, key=sum_dict.get)
  max_value = sum_dict[max_key]
  
  print(f"인구가 가장 많은 연령대: {max_key}")
else:
  print("숫자 데이터가 없습니다.")
```

9. '총 인구수' 라는 키를 추가하고, 해당 키에 0세부터 24세까지의 인구 합을 데이터로 하여 population 데이터에 추가하시오.

```python
# 9번 문제 답안 작성

age_keys = [key for key in population.keys() if key != '행정기관']

total_population = [sum(counts) for counts in zip(*(population[k] for k in age_keys))]

population['총 인구수'] = total_population

population
```
